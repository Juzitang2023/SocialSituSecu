```python
import pandas as pd
df = pd.read_csv('../../data/restaurant/restaurant_truth2.csv',
                names=['name', 'BUILDING', 'STREET', 'ZIPCODE', 'PHONE', 'ID'])
```

```python
row_num = df.shape[0]
```

```python
def cmp(x, y):
    if  x > y:
        return 1
    elif x < y:
        return -1
    else:
        return 0
```

```python
for i in range(row_num):
    print('-'*30,i,df.iloc[i].values.tolist())
    for j in range(row_num):
        if i== j:
            continue
        my_list = [cmp(df.iloc[i][n],df.iloc[j][n]) for n in range(5)]
        print(my_list)
```
